```python
#Read the CSV Data
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic=pd.read_csv(url)
titanic.head(100)
```

1.Create a Pie chart presenting Male and Female Propotion

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(10,6))
colors = ['lightskyblue', 'lightcoral']
plt.pie(titanic.sex.value_counts(),colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.xlabel('Female')
plt.ylabel('Male')
```

2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
titanic.isnull().sum()
```

```python
titanic.fare.fillna(method='ffill',inplace=True)
titanic.age.fillna(method='ffill',inplace=True)
titanic.sex.fillna(method='ffill',inplace=True)
titanic.isnull().sum()
```

```python
fig = plt.figure(figsize=(18,6))
colors = ['green', 'blue']
plt.scatter(titanic['age'], titanic['fare'],alpha=0.5, c=colors)
plt.xlabel('Age')
plt.ylabel('Fare')
```
